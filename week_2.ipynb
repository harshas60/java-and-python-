Program to calculate the square root of a number

```python
import math

n = float(input("Enter number: "))

if n < 0:
    print("Square root of negative numbers are imaginary")
else: 
    print("Square root of number is:",math.sqrt(n))
```

Program to calculate area and perimeter of triangle and circle

```python
# area and perimeter of triangle and circle

pi = 3.1415

# triangle

base_triangle = float(input("Enter triangle base: "))
side_2 = float(input("Enter triangle side2: "))
side_3 = float(input("Enter triangle side3: "))
height_triangle = float(input("Enter triangle height "))
area_triangle = (base_triangle * height_triangle) / 2

perimeter_triangle = base_triangle + side_2 + side_3

# circle

radius_circle = float(input("Enter circle radius: "))
area_circle = pi * (radius_circle ** 2)
perimeter_circle = 2 * pi * radius_circle

# print results

print("Area of Triangle:", area_triangle)
print("Perimeter of Triangle:", perimeter_triangle)
print("Area of Circle:", area_circle)
print("Perimeter of Circle:", perimeter_circle)
```

Program to calculate the roots of a quadratic equation

```python
# to calculate quadratic equations roots

print("The format for a quadratic equation is: \n ax^2 + bx + c = 0")

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter c: "))

discriminant = (b**2) - (4*a*c)

if discriminant < 0:
    print("Imaginary roots:")
# calculate two roots
root1 = (-b - discriminant**0.5) / (2*a)
root2 = (-b + discriminant**0.5) / (2*a)

print("Roots of the equation are:",root1,"and", root2)
```

Program to swap to variables

```python
a = 10
b = 20

print(f"Before swapping: ({a},{b})")

a,b = b,a

print(f"After swapping: ({a},{b})")
```

Program to convert degrees celsius to farenheit

```python
celsius = float(input("Enter temperature in degrees celsius: "))
farenheit = (celsius * (9/5)) + 32

print("Temperature in degrees farenheit =", farenheit)
```

Program to convert kilometers to miles

```python
conversion_factor = 0.621371

kilometers = float(input("Enter km: "))
miles = kilometers * conversion_factor

print(kilometers,"km =",miles,"miles")
```
