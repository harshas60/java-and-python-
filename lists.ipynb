```python
import random
print("welcome to the rock paper scissors game")
print(" 0 = rock")
print(" 1 = paper")
print(" 2 = scissors")
try:
    people_choose = int(input("what do u choose? type 0 for rock, 1 for paper or 2 for scissors: "))
except ValueError:
    print("you typed an invalid number")
else:
    if people_choose not in [0, 1, 2]:
        print("you typed an invalid number")
    else:
        computer_choice = random.randint(0,2)
        print("you:", people_choose)
        print("computer:", computer_choice)
        if people_choose == computer_choice:
            print("its a draw")
        elif (people_choose == 0 and computer_choice == 2) or (people_choose == 1 and computer_choice == 0) or (people_choose == 2 and computer_choice == 1):
            print("people wins")
            print("computer loses")
        else:
            print("computer wins")
            print("you lose")
```
